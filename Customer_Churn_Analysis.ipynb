## **Exploratory Data Analysis:**

### **What % of customers are still active vs churned.**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Churn distribution
churn_counts = df['Attrition_Flag'].value_counts(normalize=True) * 100
print(churn_counts)

sns.countplot(x='Attrition_Flag', data=df)
plt.title("Customer Attrition Distribution")
plt.show()
```

### **Demographics Analysis**

**Distribution By age:**

```python

sns.histplot(df['Customer_Age'], bins=20, kde=True)
plt.title("Age Distribution of Customers")
plt.show()

```

**Distribution By Gender:**

```python


sns.countplot(x='Gender', hue='Attrition_Flag', data=df)
plt.title("Churn by Gender")
plt.show()

```

**Distribution By Education:**

```python
sns.countplot(x='Education_Level', hue='Attrition_Flag', data=df)
plt.title("Churn by Education Level")
plt.xticks(rotation=45)
plt.show()

```

**Distribution By Income:**

```python
sns.countplot(x='Income_Category', hue='Attrition_Flag', data=df)
plt.title("Churn by Income Category")
plt.xticks(rotation=45)
plt.show()
```

## **Behavior & Credit Utilization**

**Utilization Ratio:**

```python

sns.histplot(df['Avg_Utilization_Ratio'], bins=20, kde=True)
plt.title("Average Credit Utilization Ratio")
plt.show()


```

**Transactions Vs Amount**

```python

sns.scatterplot(x='Total_Trans_Ct', y='Total_Trans_Amt', hue='Attrition_Flag', data=df)
plt.title("Transactions vs Amount by Churn")
plt.show()

```

**Inactivity Vs Churn Risk**

```python


sns.scatterplot(x='Months_Inactive_12_mon', y='Contacts_Count_12_mon', hue='Attrition_Flag', data=df)
plt.title("Inactivity vs Support Contacts (Churn Risk)")
plt.show()
```
